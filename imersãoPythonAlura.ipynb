```python
# Create a dictionary to map English column names to Portuguese
column_translation = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Rename the columns using the dictionary
df = df.rename(columns=column_translation)

# Display the new column names to confirm the translation
print(df.columns)
```

```python
# Create a dictionary to map English experience level abbreviations to Portuguese words
senioridade_translation = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Replace the values in the 'senioridade' column using the dictionary
df['senioridade'] = df['senioridade'].replace(senioridade_translation)

# Display the updated value counts to confirm the translation
print(df['senioridade'].value_counts())
```

```python
# Create a dictionary to map English employment type abbreviations to Portuguese words
contrato_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Replace the values in the 'contrato' column using the dictionary
df['contrato'] = df['contrato'].replace(contrato_translation)

# Display the updated value counts to confirm the translation
print(df['contrato'].value_counts())
```

```python
# Create a dictionary to map English company size abbreviations to Portuguese words
tamanho_empresa_translation = {
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Replace the values in the 'tamanho_empresa' column using the dictionary
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_translation)

# Display the updated value counts to confirm the translation
print(df['tamanho_empresa'].value_counts())
```

```python
# Create a dictionary to map remote ratio values to Portuguese words
remoto_translation = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}

# Replace the values in the 'remoto' column using the dictionary
df['remoto'] = df['remoto'].replace(remoto_translation)

# Display the updated value counts to confirm the translation
print(df['remoto'].value_counts())
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome' : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salario' : [4000, np.nan, 5000, np.nan, 100000]
})

# Preencher com a mdia salarial
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preencher com a mediana salarial
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dias' : ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    'Temperaturas' : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_Ffill"] = df_temperaturas['Temperaturas'].ffill()

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dias' : ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    'Temperaturas' : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_Bfill"] = df_temperaturas['Temperaturas'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome' : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'cidades' : ["So Paulo", np.nan, "Curitiba", np.nan, "Belm"]
})

df_cidades["cidade_preenchida"] = df_cidades['cidades'].fillna("No Informado")
df_cidades
```
